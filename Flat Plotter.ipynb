```python
from pylab import *
```

```python
# Set output directory name.
dir_name = "output/"
```

```python
# Read files.
r = np.genfromtxt(dir_name + "r.asc")
z = np.genfromtxt(dir_name + "z.asc")
s = np.genfromtxt(dir_name + "s.asc")
f = np.genfromtxt(dir_name + "f.asc")
u = np.genfromtxt(dir_name + "u.asc")
res = np.genfromtxt(dir_name + "res.asc")
```

```python
# Array shape.
print np.shape(u)
```

```python
# Get spatial steps.
dr, dz = (r[1,0] - r[0,0]), (z[0,1] - z[0,0])
print dr, dz
```

```python
# Get grid extension for imshow.
r_min, r_max = np.min(r), np.max(r)
z_min, z_max = np.min(z), np.max(z)
```

```python
# Simple plotter.
def plot_2d(u, name):
    fig, ax = plt.subplots()
    im = ax.imshow(u.T, origin = 'lower', extent = [r_min, r_max ,z_min, z_max])
    ax.set_xlabel(r"$\rho$")
    ax.set_ylabel(r"$z$")
    ax.set_title(name)
    fig.colorbar(im, ax = ax)
    plt.show()
```

```python
plot_2d(r, r"$\rho$ grid")
```

```python
plot_2d(z, r"$z$ grid")
```

```python
plot_2d(s, r"Linear source $s$")
```

```python
plot_2d(f, r"RHS $f$")
```

```python
plot_2d(u, r"Solution $u$")
```

```python
# Examine interior points of the residual.
plot_2d(res[:-1, :-1], r"Residual $r$")
```
